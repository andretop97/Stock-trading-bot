# Avaliação de ação da bolsa


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
from sklearn.model_selection import train_test_split
```

### Tratandos os dados

Importando dados


```python
wdown = pd.read_csv(".\data\WDO$N_M1.csv", sep="\t")
winwn = pd.read_csv(".\data\WIN$N_M1.csv", sep="\t")
```

Tratando os dados: filtrando as colunas


```python
date_time = [datetime.strptime(f'{data["<DATE>"]} {data["<TIME>"]}',
                               '%Y.%m.%d %H:%M:%S') for index, data in wdown.iterrows()]
wdown["<DATETIME>"] = date_time
wdown.set_index("<DATETIME>", inplace=True)
wdown = wdown.drop(columns=['<DATE>', '<TIME>'])

date_time = [
    datetime.strptime(f'{data["<DATE>"]} {data["<TIME>"]}',  '%Y.%m.%d %H:%M:%S') for index, data in winwn.iterrows()]
winwn["<DATETIME>"] = date_time
winwn.set_index("<DATETIME>", inplace=True)

winwn = winwn.drop(columns=['<DATE>', '<TIME>'])

del date_time
```

```python
wdown_monthly = wdown.resample('BM').mean()
wdown_daily = wdown.resample('D').mean()
```

Variação diaria e mensal

```python
daily_pct_change = wdown_daily["<CLOSE>"].pct_change()
monthly_pct_change = wdown_monthly["<CLOSE>"].pct_change()

daily_pct_change.fillna(0, inplace=True)
monthly_pct_change.fillna(0, inplace=True)
```

Variação acumulada diaria e mensal

```python
cum_daily_return = (1 + daily_pct_change).cumprod()
cum_monthly_return = (1 + monthly_pct_change).cumprod()
```

Variação acumulada diaria e mensal

```python
cum_daily_return.plot()
cum_monthly_return.plot()
plt.axhline(y = 1, color = 'r', linestyle = '-') 
```

Criando coluna de decisão -1 (Compra), 0 (Mantem), 1 (Venda).

```python
ptc_change =  wdown["<CLOSE>"].pct_change()
ptc_change.fillna(0, inplace=True)


wdown["<PTC_CHANGE>"] = ptc_change

wdown["<SIGNAL>"] = np.where(wdown["<PTC_CHANGE>"] < -0.0001, 1, 0)
wdown["<SIGNAL>"] =  np.where(wdown["<PTC_CHANGE>"] > 0.0001, -1, wdown["<SIGNAL>"])
```

Criando dados de treino e teste

```python
x = wdown.iloc[:,:-1]
y = wdown.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(x, y)
```
